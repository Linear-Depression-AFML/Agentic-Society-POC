```python
import sender_reciever_1v1comms as sr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame(columns=["Game", "Start Sender Amount", "Start Receiver Amount", "Amount Sent By Sender", "Amount Sent By Receiver", "Final Sender Amount", "Final Receiver Amount", "Sender Promise", "Receiver Promise"])
s_wins = 0
r_wins = 0
ties = 0
game_fails = 0
rounds = 10
for i in range(50):
    try:
        result = sr.run_game(15, i+1, df, rounds)
        if result == 1:
            s_wins += 1
        elif result == -1:
            r_wins += 1
        else:
            ties += 1
    except Exception as e:
        print(f"Error in game {i+1}: {e}")
        game_fails += 1
        
    print(df.tail(10))
    
df.to_csv("sr_games_15m_10r.csv", index=False)
```

```python
df = pd.read_csv("sr_games_15m_10r.csv")
```

```python
print("Sender wins:", s_wins, "Receiver wins:", r_wins, "Ties:", ties, "Game fails:", game_fails)
```

```python
for i in range(0, len(df), 10):
    new_df = df.iloc[i:i+10]
    new_df = new_df.reset_index(drop=True)
    plt.plot(new_df["Final Sender Amount"])
    plt.xticks(range(10))
plt.title("Final Sender Amount over 100 Games")
plt.xlabel("Round")
plt.ylabel("Sender Amount")
plt.show()
```

```python
for i in range(0, len(df), 10):
    new_df = df.iloc[i:i+10]
    new_df = new_df.reset_index(drop=True)
    plt.plot(new_df["Final Receiver Amount"])
    plt.xticks(range(10))
plt.title("Final Receiver Amount over 100 Games")
plt.xlabel("Round")
plt.ylabel("Receiver Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(10, 7))
plt.scatter(df["Amount Sent By Sender"], df["Amount Sent By Receiver"], alpha=0.5)

plt.title("Amount Sent by Receiver vs. Amount Sent by Sender (All Games)")
plt.xlabel("Amount Sent By Sender")
plt.ylabel("Amount Sent By Receiver")
plt.grid(True)
plt.show()
```
